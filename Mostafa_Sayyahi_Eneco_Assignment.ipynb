## Q1.1

```python
countries_by_airports = countries.select(['code','name']).join(
    airports.group_by('iso_country').agg(
        pl.col('id').count().alias('airport_count')
    )    
    , how= 'left'
    , left_on =  'code'
    , right_on = 'iso_country'
).filter(~ ((pl.col('code') == 'ZZ') & pl.col('airport_count').is_null()) )\
.with_columns(
    pl.col("airport_count").fill_null(0)
)
```

```python
top_filter = pl.col("airport_count").rank(method="min",descending=True) <=3
countries_by_airports.filter(top_filter).sort('airport_count', descending = True)
```

```python
pl.Config.set_tbl_rows(-1)
bottom_filter =  pl.col("airport_count").rank(method="min",descending=False) <= 10
countries_by_airports.filter(bottom_filter).sort('airport_count', descending = True)
```

## Q1.2

```python
longest_runways_per_airport = runways.select(['airport_ref', 'length_ft', 'width_ft']).filter(
    pl.col("length_ft")
    .rank(method="min", descending=True)
    .over("airport_ref") == 1
)

```

```python
longest_runways_per_country=airports.select(['id','name','iso_country'])\
.join(
    longest_runways_per_airport
    , how = 'inner'
    , left_on = 'id'
    , right_on = 'airport_ref'
).filter(
    pl.col("length_ft")
    .rank(method="min", descending=True)
    .over("iso_country") == 1
).sort('length_ft', descending = True)
    
```

```python
import psycopg2
conn = psycopg2.connect(host="code001.ecsbdp.com", port=5432, dbname="chinook", user="larry", password="iddqd")
pl.read_database(
    """
    WITH track_type AS (
        SELECT
            track_id
            , CASE WHEN g.name = 'Jazz' THEN 1 ELSE 0 END AS is_jazz
        FROM track AS t
        LEFT JOIN genre AS g USING (genre_id)
    )
    , customer_sales_by_type AS (
        SELECT
            customer_id
            , MAX(is_jazz) AS is_jazz_customer
            , SUM(total) total_sales
        FROM invoice AS i
        LEFT JOIN invoice_line AS il USING(invoice_id)
        LEFT JOIN track_type AS tt USING(track_id)
        GROUP BY customer_id
    )
    SELECT 
        is_jazz_customer
        , SUM(total_sales) AS total_sales
    FROM customer_sales_by_type
    GROUP BY is_jazz_customer
    """
    , conn
)
```
