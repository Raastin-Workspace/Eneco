```python
countries_by_airports = countries.select(['code','name']).join(
    airports.group_by('iso_country').agg(
        pl.col('id').count().alias('airport_count')
    )    
    , how= 'left'
    , left_on =  'code'
    , right_on = 'iso_country'
).filter(~ ((pl.col('code') == 'ZZ') & pl.col('airport_count').is_null()) )\
.with_columns(
    pl.col("airport_count").fill_null(0)
)
```

```python
top_filter = pl.col("airport_count").rank(method="min",descending=True) <=3
countries_by_airports_top3 = countries_by_airports.filter(top_filter).sort('airport_count', descending = True)

countries_by_airports_top3.write_csv('./results/q1_countries_by_airports_top3.csv')
countries_by_airports_top3
```

```python
pl.Config.set_tbl_rows(-1)
bottom_filter =  pl.col("airport_count").rank(method="min",descending=False) <= 10
countries_by_airports_bottom10 = countries_by_airports.filter(bottom_filter).sort('airport_count', descending = True)
countries_by_airports_bottom10.write_csv('./results/q1_countries_by_airports_bottom10.csv')
countries_by_airports_bottom10
```

## Q1.2

```python
# fixed min -> ordinal to avoid duplicated airports
longest_runways_per_airport = runways.select(['airport_ref', 'length_ft', 'width_ft']).filter(
    pl.col("length_ft")
    .rank(method="ordinal", descending=True)
    .over("airport_ref") == 1
)

```

```python
longest_runways_per_country=airports.select(['id','name','iso_country'])\
.join(
    longest_runways_per_airport
    , how = 'inner'
    , left_on = 'id'
    , right_on = 'airport_ref'
).filter(
    pl.col("length_ft")
    .rank(method="min", descending=True)
    .over("iso_country") == 1
).sort('length_ft', descending = True)
    
```

```python
import requests as req
class api:
    def __init__(self, domain, timeout= (5,30)):
        self.domain = domain
        self.timeout = timeout
        self.session = req.Session()
    
    def _url(self, url_path):
        return self.domain + url_path
        
    def get(self, url_path, params = None, headers = None):
        url = self._url(url_path)
        print('***started***')
        print(f'''url : {url}''')
        try:
            res = self.session.get(url, params = params, headers = headers, timeout = self.timeout)
            # res.raise_for_status()
            
        except Exception as e:
            print(f"Type: {e.__class__.__name__}")
            print(f"Full: {type(e)}")
            print(f"Message: {str(e)}") 
            # raise
        else:
            print('successful')
        finally:            
            print(f'''***finished with staus code: {res.status_code}***''')
            print(f'''***finished with reason: {res.reason}***\n''')
            return res


    def put(self, url_path, data, params = None, headers = None):
        url = self._url(url_path)
        print('***started***')
        print(f'''url : {url}''')
        try:
            res = self.session.put(url, data = data , params = params, headers = headers, timeout = self.timeout)
            # res.raise_for_status()
            
        except Exception as e:
            print(f"Type: {e.__class__.__name__}")
            print(f"Full: {type(e)}")
            print(f"Message: {str(e)}") 
            # raise
        else:
            print('successful')
        finally:            
            print(f'''***finished with staus code: {res.status_code}***''')
            print(f'''***finished with reason: {res.reason}***\n''')
            return res

    def post(self, url_path, data = None, files = None, params = None, headers = None):
        url = self._url(url_path)
        print('***started***')
        print(f'''url : {url}''')
        try:
            res = self.session.post(url, data = data , params = params, headers = headers, timeout = self.timeout)
            # res.raise_for_status()
            
        except Exception as e:
            print(f"Type: {e.__class__.__name__}")
            print(f"Full: {type(e)}")
            print(f"Message: {str(e)}") 
            # raise
        else:
            print('successful')
        finally:            
            print(f'''***finished with staus code: {res.status_code}***''')
            print(f'''***finished with reason: {res.reason}***\n''')
            return res
        
```
